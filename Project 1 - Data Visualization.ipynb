### 2018 Dataframe

```python
df_2018 = pd.read_csv('./2018.csv')
```

```python
df_2018 = df_2018.drop('Unnamed: 27', axis = 1)
```

```python
df_2018.head()
```

```python
df_2018.shape
```

```python
column_list = df_2018.columns.values.tolist()
column_list
```

```python
df_2018.hist(column='CRS_DEP_TIME');
```

```python
df_2018.OP_CARRIER.unique()
```

```python
df_2018 = df_2018[df_2018.OP_CARRIER != 'VX']
```

```python
df_2018.OP_CARRIER.unique()
```

```python
df_2018 = df_2018[df_2018.WEATHER_DELAY.isna() == False]
df_2018 = df_2018.reset_index(drop = True)
df_2018.head()
```

```python
df_2018['CRS_DEP_TIME'] = df_2018['CRS_DEP_TIME'].astype(float)
df_2018.head()
```
